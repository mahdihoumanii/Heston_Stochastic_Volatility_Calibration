# Calibration Report Snippets
Lightweight notebook producing key figures for the report.

```python
import json
import matplotlib.pyplot as plt
import pandas as pd

with open('../reports/calibration_result.json', 'r') as f:
    result = json.load(f)
params = result['params']
print('Loaded calibration result:', params)

param_df = pd.DataFrame(list(params.items()), columns=['Parameter','Value'])
param_df.plot.bar(x='Parameter', y='Value', legend=False, figsize=(5,4))
plt.title('Calibrated parameters')
plt.tight_layout()
plt.savefig('../reports/figures/calibrated_params_bar.png', dpi=150)
plt.show()
```
